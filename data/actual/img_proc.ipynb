```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#img = cv2.imread('1a.jpg',)
img = cv2.imread('5d.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(img, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(cv2.medianBlur(img, 31), cmap='gray')
plt.show()
```

```python
plt.hist(img.ravel(), bins=255)
plt.show()
```

```python
clahe = cv2.createCLAHE(clipLimit=5.0, )#tileGridSize=(8,8))
cl1 = clahe.apply(img)
```

```python

plt.figure(figsize=(15, 15))
plt.imshow(cl1, cmap='gray')
plt.show()
```
